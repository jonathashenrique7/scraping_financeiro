```python
import pandas as pd 
import requests as rq 

# Digite "my user agent" no buscador do Google, copie e cole seu user-agent Mozilla como valor da variável "header"
header = {''}

url = 'https://br.investing.com/equities/Brazil'

request = rq.get(url, headers=header)

# Utilize a função read_html() + pandas em sites onde os dados estão em formato de tabelas 
df = pd.read_html(request.text)
df
```

```python
for i , tabela in enumerate(df):
    print('-------------')
    print(i)
    print(tabela)
```

```python
# Tansformando os dados em um dataframe organizado 

df = pd.DataFrame(df[0])
df

# Isso deve retornar as linhas com índices entre 4 e 80 (ou seja, excluindo 81).
```

```python
df['Último'] = df['Último'] / 100 
df['Máxima'] = df['Máxima'] / 100
df['Mínima'] = df['Mínima'] / 100 

df
```
